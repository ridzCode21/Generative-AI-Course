```python
List  = ['riddhesh' , 23 , 60.55 , True]
List[-2] = 100.25
type(List)
print(List)
```

```python
print(List[0])
print(List[:4])
print(List[1:3])
print(List[-1])
print(List[-3:-1])
print(List[::2])
```

```python
List.insert(List.index(23),28)
print(List.index(23))
print(List)
```

```python
## Iterating with index
for index,number in enumerate(List):
    print(index,number)
```

List Comprehension
Basics Syantax [expression for item in iterable]

with conditional logic [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]

```python
list1 = [1,2,3,4,5]
list2 = list1[::-1]
print(list2)

[print(item,j) for item in list1 for j in list1] 
```

```python
# Example 1: Square of each number in list1
squares = [x**2 for x in list1]
print("Squares:", squares)

# Example 2: Filter even numbers from list1
evens = [x for x in list1 if x % 2 == 0]
print("Even numbers:", evens)

# Example 3: Create a list of tuples (number, square)
tuples = [(x, x**2) for x in list1]
print("Tuples (number, square):", tuples)

# Example 4: Flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [num for row in matrix for num in row]
print("Flattened list:", flattened)

# Example 5: Conditional logic - Replace odd numbers with 'Odd'
replace_odds = [x if x % 2 == 0 else 'Odd' for x in list1]
print("Replace odds:", replace_odds)
```
