Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.


```python
arr = [1,2,3]
arr  = tuple(arr)
print(type(arr))
print(arr)
```

```python
list1 = ['apple' , 'banana' , 'cherry']
list1  = tuple(list1)
list2 =  list1*3
print(list2)
list1.count('apple')
```

## packing and unpacking 


```python
t = (1, "apple", 3.5)
val1, val2, val3 = t
print(val1)
print(val2)
print(val3)
```

```python
# 5. Unpacking in Loops

# Useful when iterating over sequences of tuples:
pairs = [(1, 'a'), (2, 'b'), (3, 'c')]

# for num, char in pairs:
#     print(num, char)
[print(num, char) for num, char in pairs]
```

```python
def unpack_tuple(tup):
    a, *b, c = tup
    return a, b, c

a,b,c = unpack_tuple((1, 2, 3, 4, 5))
print(a,b,c)  # Output: (1, [2, 3, 4],
```
