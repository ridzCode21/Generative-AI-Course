# Using else and finally
# The else block runs only if no exception occurred in the try, and finally always runs whether an exception happened or not.

# else is useful for code that should run only when everything in try succeeded.

# finally is useful for cleanup tasks like closing files, releasing locks, or closing database connections.



```python
try:
    f = open("data.txt", "r")
    content = f.read()
except FileNotFoundError:
    print("File not found.")
else:
    print("File read successfully.")
finally:
    # This will run whether or not an exception occurred
    try:
        f.close()
    except NameError:
        pass
```
